```python
import folium
import pandas as pd

# Charger le fichier CSV avec les escales
file_path = "Relever-Nautilus-Position_UTF82.csv"  # Assure-toi que le fichier est bien √† cet emplacement
df = pd.read_csv(file_path, delimiter=";", encoding="utf-8")

# Cr√©er une carte centr√©e sur un point moyen des donn√©es
m = folium.Map(location=[0, 0], zoom_start=2)

# Ajouter les escales du Nautilus sur la carte
for index, row in df.iterrows():
    folium.Marker(
        location=[row["Latitude_D√©cimal"], row["Longitude_D√©cimal"]],
        popup=f"{row['Escales du Nautilus']} ({row['Date']})",
        tooltip=row["Escales du Nautilus"]
    ).add_to(m)

# Tracer le trajet avec une ligne reliant les points
coordinates = list(zip(df["Latitude_D√©cimal"], df["Longitude_D√©cimal"]))
folium.PolyLine(locations=coordinates, color="blue", weight=2.5, opacity=1).add_to(m)

# Afficher la carte interactive
m
```
